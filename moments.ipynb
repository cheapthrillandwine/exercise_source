```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
img = cv2.imread("gauge-15.jpg")
imgg = cv2.medianBlur(img, 5)
gray = cv2.cvtColor(imgg, cv2.COLOR_RGB2GRAY)
_, contours, _ = cv2.findContours(gray, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

maxCont=contours[0]
for c in contours:
    if len(maxCont)<len(c):
        maxCont = c
mu = cv2.moments(maxCont)
x,y = int(mu["m10"]/mu["m00"]), int(mu["m01"]/mu["m00"])

cv2.circle(gray, (x,y),4,100,2,4)
plt.imshow(gray)
plt.colorbar()
plt.show()

print(x,y)
```
